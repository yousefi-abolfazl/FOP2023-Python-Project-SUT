<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        ابتدا دامنه تغییرات ستون Price را محاسبه کنید و آن را به 4 بازه مساوی با نام‌های cheap , AveUnderMean , AveUpperMean , Expensive تقسیم کنید. به هر بازه تعداد خانه‌هایی که قیمتشان در این بازه قرار دارد را نسبت دهید و در آخر نمودار countplot را برای این بازه‌ها رسم کنید.
    </font>
</div>

```python
range = df['Price'].max() - df['Price'].min()
bins = [df['Price'].min() - 1, range / 4, range / 2, 3 * range / 4, df['Price'].max() + 1]
labels = ['cheap', 'AveUnderMean', 'AveUpperMean', 'Expensive']
df['PriceRange'] = pd.cut(df['Price'], bins=bins, labels=labels)
sns.countplot(data=df, x='PriceRange', order=labels)
plt.xlabel('Price Range')  
plt.ylabel('Count')
plt.title('Count Plot for Price Ranges') 
plt.show()
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        ابتدا 2 درصد از داده‌های ستون Price را به صورت تصادفی انتخاب کنید و سپس نمودار خط‌ شکسته را برای این داده‌ها رسم کنید.(محورها دارای label باشند و رنگ نمودار قرمز باشد.)
    </font>
</div>

```python
sample_size = int(0.02 * len(df))
random_indices = np.random.choice(df.index, size=sample_size, replace=False)
sample_data = df.loc[random_indices, 'Price']
plt.plot(sample_data.index, sample_data, marker='o', linestyle='-', color='red')
plt.title('A step plot for the top 2% of the Price column data.')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        همانطور که میبینید نمودار قبلی ظاهر خوبی ندارد و شرط تابع بودن را نقض کرده است، داده‌های تصادفی مرحله قبل را به گونه‌ای sort کنید نمودار ما تبدیل به یک تابعی شود که اکیدا یکنوا نیست و نوسان دارد.
    </font>
</div>

```python
sorted_data = sample_data.sort_values()
plt.plot(sorted_data, sorted_data.index, linestyle='-', color='red')
plt.title('A step plot for the top 2% of the Price column data (oscillation).')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(df['Area'], df['Price(USD)'], color='b')
plt.xlabel('Area')  
plt.ylabel('Price(USD)')
plt.title('Line Plot of Price(USD) vs Area')
plt.show()
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        به نمودار رسم شده در قسمت قبل، نمودار روند تغییرات مرتبه اول(خطی) را اضافه کنید.(برای پیدا کردن ضرایب رگرسیون می‌توانید از تابع polyfit در کتابخانه numpy استفاده کنید.)
    </font>
</div>

```python
df1 = df.dropna()
coefficients = np.polyfit(df1['Area'], df1['Price'], 1)
plt.plot(df1['Area'], np.polyval(coefficients, df1['Area']), color='blue', linestyle='-', label='Linear Trend (polyfit)')
plt.xlabel('Area')
plt.ylabel('Price (USD)')
plt.title('Broken Line Chart with Linear Trend (polyfit) of Price based on Area')
plt.legend()
plt.show()
```

<div dir = "rtl" style="font-face:IranSansX;direction:rtl;line-height:250%;font-size:small;">
	<font face="IranSansX" size=3>
        نمودار روند تغییرات خطی که در قسمت قبل رسم کردیم را به صورت خط‌چین با رنگ قرمز رسم کنید.
    </font>
</div>

```python
df1 = df.dropna()
coefficients = np.polyfit(df1['Area'], df1['Price'], 1)
plt.plot(df1['Area'], np.polyval(coefficients, df1['Area']), color='r', linestyle='--', label='Linear Trend (polyfit)')
plt.xlabel('Area')
plt.ylabel('Price (USD)')
plt.title('Broken Line Chart with Linear Trend (polyfit) of Price based on Area')
plt.legend()
plt.show()
```
